```python
# SHOW COLUMNS
df.columns
```

```python
# COLUMN RENAME
df =df.rename(columns={'Order Id':'order_id','Order Date':'order_date','Ship Mode':'ship_mode','Postal Code':'postal_code','Sub Category':'product_name','Product Id':'product_id','cost price':'cost_price','List Price':'list_price','Discount Percent':'discount_percent','Segment':'segment','Country':'country','City':'city','State':'state','Region':'region','Category':'category','Quantity':'quantity'})
df
```

```python
# CHECK DATA TYPES OF COLUMNS
df.dtypes
```

```python
# CHANGE DATE TYPE
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
# Extract year, month, and day
df['Year'] = df['order_date'].dt.year
df['Month'] = df['order_date'].dt.month
df['Day'] = df['order_date'].dt.day
```

```python
df.dtypes
```

```python
# CHECK NULL VALUES
df.isnull().sum()
```

```python
#FILL NA
df=df.fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
print(df['ship_mode'].value_counts())
df
```

```python
# CALCULATE DISCOUNT
df['discount']=df['list_price']*(df['discount_percent']/100)
```

```python
# CALCULATE SALES PRICE
df['sales_price']=df['list_price']-df['discount']
```

```python
# CALCULATE PROFIT
df['profit']=df['sales_price']-df['cost_price']
```

```python
df.columns
```

```python
# REMOVE WHITE SPACES
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x) 

print(df)
```
